```python
x = df.drop(['Class'], axis=1)
y = df['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
xData = x.values
yData = y.values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(xData, yData, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rfc = RandomForestClassifier()
Rfc.fit(xtrain, ytrain)
yprediction = Rfc.predict(xtest)
```

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, matthews_corrcoef
```

```python
n_outlier = len(fraud_card)
n_error = (ytest != yprediction).sum()
```

```python
print("Used Random Forest Classifier.")
accuracy = accuracy_score(ytest, yprediction)
print("The Acuracy is: ", accuracy)

precision = precision_score(ytest, yprediction)
print("The Precision is: ", precision)

f1_scre = f1_score(ytest, yprediction)
print("The F1 Score is: ", f1_scre)

MCC = matthews_corrcoef(ytest, yprediction)
print("The Matthews Correlation Coefficient is: ", MCC)
```

```python
labels = ['Valid', 'Fraud']
conf_metrcs = confusion_matrix(ytest, yprediction)
plt.figure(figsize=(12,12))
sns.heatmap(conf_metrcs, xticklabels=labels, yticklabels=labels, annot=True, fmt="d")
plt.title("Confusion Matrix")
plt.ylabel('True Class')
plt.xlabel('Predicted Class')
plt.show()
```
